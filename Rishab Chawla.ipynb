# 3.Plot how data field values are distributed across the locations 'Argentina-Buenos_Aires', 'Argentina-CABA', 'Argentina-Catamarca'

```python
t=data[["report_date","location","data_field","value"]]
```

```python
t.drop('report_date',axis=1,inplace=True)
```

```python
t2=t[t.location=='Argentina-CABA']
t1=t[t.location=='Argentina-Buenos_Aires']
t3=t[t.location=='Argentina-Catamarca']
t3
```

```python
y1=t1.groupby('data_field').sum()
y2=t2.groupby('data_field').sum()
y3=t3.groupby('data_field').sum()
y3
```

```python
x1=t1.data_field.unique()
x2=t2.data_field.unique()
x3=t3.data_field.unique()
```

```python
fig = plt.figure(1, figsize=(30,15))
plt.plot(x1,y1)
plt.plot(x2,y2)
plt.plot(x3,y3)
```

```python
fig,axes=plt.subplots()
labels=x1
fig = plt.figure(1, figsize=(80,80))
#value=y1
#explode = (0, 0.1, 0, 0,0,0)
plt.title("Argentina-Buenos_Aires",bbox={'facecolor':'3.8', 'pad':5})
plt.rcParams['font.size'] = 10.0
axes.axis('equal')  
plt.tight_layout()
axes.pie(y1,autopct='%1.2f', startangle=90 ,shadow=True,labeldistance=1.2,radius=1 )
plt.legend(labels, bbox_to_anchor=(0.85,1.025))
plt.show()

```

```python
fig,axes=plt.subplots()
labels=x2
fig = plt.figure(1, figsize=(80,80))
#value=y1
#explode = (0, 0.1, 0, 0,0,0)
plt.title("Argentina-CABA",bbox={'facecolor':'3.8', 'pad':5})
plt.rcParams['font.size'] = 10.0
axes.axis('equal')  
plt.tight_layout()
axes.pie(y2,autopct='%1.2f', startangle=90 ,shadow=True,labeldistance=1.2,radius=1 )
plt.legend(labels, bbox_to_anchor=(0.85,1.025))
plt.show()
```

```python
fig,axes=plt.subplots()
labels=x3
fig = plt.figure(1, figsize=(80,80))
#value=y1
#explode = (0, 0.1, 0, 0,0,0)
plt.title("Argentina-Catamarca",bbox={'facecolor':'3.8', 'pad':5})
plt.rcParams['font.size'] = 10.0
axes.axis('equal')  
plt.tight_layout()
axes.pie(y3,autopct='%1.2f', startangle=90 ,shadow=True,labeldistance=1.2,radius=1 )
plt.legend(labels, bbox_to_anchor=(0.85,1.025))
plt.show()
```

```python
query=data.query("location=='Argentina-Buenos_Aires' | location=='Argentina-CABA' | location=='Argentina-Catamarca' ")
newData=query.groupby(['data_field','location'],as_index=False)['value'].sum()
# newData = newData[newData['value'] != 0]
l = list(newData.data_field.unique())
newData.set_index('data_field',inplace=True)
newData
```

```python
newData.loc[l[1]]['value']
```

```python
fig,axes=plt.subplots(nrows=2,ncols=3,figsize=(20,18))
k=0
for i in range(2):
    for j in range(3):
        temp=newData.loc[l[k]]
        axes[i][j].pie(temp['value'],autopct="%1.1f%%", startangle=90)
        axes[i][j].title.set_text(l[k])
        axes[i][j].axis('equal')
        k+=1
plt.legend(labels=temp['location'], bbox_to_anchor=(1.0,0.0) ,fontsize='xx-large')
plt.tight_layout(pad=5)
```

# 4.Build a program to predict the correct word for wrongly typed word. for example whst => what, numbr =>number, good => good

```python
from difflib import get_close_matches 

def closeMatches(patterns, word): 
     print("Matching pattern words are:" , get_close_matches(word, patterns)) 

with open('big.txt')as file:
    data=file.read()
    #print(data)
    ls=[]
    for word in data.split():
        ls.append(word)
    file.close()
    #print(ls)
with open('words.txt')as file:
    data=file.read()
    #print(data)
    l=[]
    for word in data.split():
        l.append(word)
    file.close()

#if w in l:
    #print('yes')
ls = list(dict.fromkeys(ls))
# Driver program 
if __name__ == "__main__": 
    word = input("Enter word: ")
    #patterns = ['ape', 'apple', 'peach', 'puppy']
    if word not in l:
        closeMatches(ls, word)
    else:
        print("Word is correct")
```
